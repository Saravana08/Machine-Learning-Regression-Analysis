```python
data['MiscFeature'].fillna('Shed',inplace=True)
```

```python
x = data[['MSSubClass','BldgType','LotArea']]
y = data['SalePrice']
```

```python
x.dtypes
```

```python
x['BldgType'].value_counts()
```

```python
x = pd.get_dummies(x)
```

```python
x
```

```python
import sklearn
x.shape
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(x,y,test_size=0.2)
```

```python
print(xtr.shape)
print(xts.shape)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression,ridge_regression
model1 = LinearRegression()  # GOt Defined
model1.fit(xtr,ytr)          # Got Trained
predictions = model1.predict(xts)   # Got Predictions
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(yts,predictions)
```

```python
mean_absolute_error(yts,predictions)
```
